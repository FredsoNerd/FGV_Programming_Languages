Seja d(n) definido como a suma dos divisores inteiros de n. Se d(a) = b e d(b) = a, sendo a ≠ b, então a e b são um "par amigável" de números.  

Por exemplo, os divisores inteiros de 220 são 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 e 110; então d(220) = 284.  
Os divisores inteiros de 284 são 1, 2, 4, 71 e 142; então d(284) = 220.  

Encontre todos os "pares amigáveis" de números menores que 20000.  Crie uma lista de tuplas com esses pares.  

```python
#seu código aqui

def soma(num):
    soma = 0
    for i in range(1, num):
        if num%i == 0:
            soma += i
            
    return soma

def find_amigaveis(limite):
    
    amigaveis = []
    amigos = []
    for n in range(1, limite):
        if n in amigos: continue
        
        soma_ = soma(n)
        if soma(soma_) == n and not n == soma_:
            amigos.append(soma_)
            
            amigaveis.append((n,soma_))
    
    return amigaveis


# VEJA QUE DESCONSIDERAMOS REPETICOES
# E PEX, (220, 284) EQUIVALE A (284, 220)
find_amigaveis(20000)
```

Usando funções geradoras, crie um objeto que emule um dado de oito faces (1-8) que seja viciado, de tal forma que a probabilidade desta função geradora retornar certo valor seja proporcional ao próprio valor (por exemplo, a face 6 é 3 vezes mais provável de sair do que a face 2).  

(a) Usando os comandos plt.plot ou plt.hist, mostre graficamente o resultado de 10000 lançamentos;  
(b) Modifique esta função geradora de forma que ela termine automaticamente quando todos os valores possíveis (1,2,3,4,5,6,7,8) tiverem sido retornados pelo menos uma vez. Neste caso, ela retornará o tempo absoluto total que ocorreu desde a primeira iteração.  

```python
#seu código aqui

def dado_d8():
    dado = []
    for i in range(1,9):
        for j in range(i):
            dado.append(i)
            
    while True:
        yield np.random.choice(dado)
```

```python
# resolucao item a)
meu_dado = dado_d8()

x = [next(meu_dado) for i in range (10000)]


plt.hist(x, 16, color='#faa0a0', label='Lancamentos dado d8')
plt.title('10000 lancamentos do d8 viciado')
plt.show()
```

Crie uma classe chamada "Elevador" que recebe, ao instanciada, o número de andares do edifício e inicia no andar mais baixo.  
Crie métodos e propriedades para permitir que o elevador:  
- Receba uma chamada (usuário aperta um botão em um andar);  
- Receba um andar como destino (usuários apertam botões dentro do elevador);  
- Informe em qual andar o elevador está;  
- Informe a sequência de andares ainda por visitar;
- Informe a quantidade de vezes que parou em cada andar (passar pelo andar sem "parar" no andar não conta);  

Leve em consideração que cada deslocamento de andar demora 3 segundos.  
Simule o comportamento do elevador atendendo à dez usuários, que têm andar de origem, andar de destino e sequência de chegada (fora e dentro do elevador) definidos aleatóriamente.  

```python
#seu código aqui

from threading import Thread

class Elevador:
    """
    Sistema de controle de elevador EMAp1000.
    
    O elevador trabalha por um sistema de filas
    de chamada, e recebe as chamadas e destinos.
    
    Quando não ha pedidos ou em emergencias, o
    elevador retorna para o terreo."""
    
    def __init__(self, andares):
        self.maxim = andares
        self.andar = 0
        self.histo = {andar:0 for andar in range(andares + 1)}
        
        self.destinos = {'Up':set(), 'Dw':set()}
        self.way = []
        
        Thread(target=self.__anda__).start()
        
    def chamar(self, andar, destino):
        if max([andar, destino]) <= self.maxim:
            if min([andar, destino]) >= 0:
                if not andar == self.andar:
                    if not destino == andar:

                        if andar > self.andar:
                            self.destinos['Up'].add(andar)
                            self.way.append(1)
                        if andar < self.andar:
                            self.destinos['Dw'].add(andar)
                            self.way.append(-1)
                        
                        if destino > andar:
                            self.way.append(1)
                            self.destinos['Up'].add(destino)
                        if destino < andar:
                            self.way.append(-1)
                            self.destinos['Dw'].add(destino)
                        
            
    def destino(self, andar):
        self.chamar(andar)
    
    def andar_atual(self):
        print('Andar atual: {}'.format(self.andar))
    
    def por_visitar(self):
        print('Visitar: {}'.format(self.destinos))
    
    def historico(self):
        print('Historico: {}'.format(self.histo))
        
    def __anda__(self):
        destino = 0
        
        while len(self.way) == 0:
            pass
        
        time.sleep(1)
        
        way = self.way[0]
        
        try:
            if way > 0:
                op = 'Up'
                destino = max(self.destinos[op])
            if way < 0:
                op = 'Dw'
                destino = min(self.destinos[op])
        except:
            self.destinos = {'Up':set(), 'Dw':set()}
            destino = 0
            way = -1
        
        while not self.andar == destino:
            time.sleep(1)
            self.andar += way
            
            if not len(self.destinos[op]) == 0:
                
                try:
                    if way > 0: destino = max(self.destinos[op])
                    if way < 0: destino = min(self.destinos[op])
                except:
                    self.destinos = {'Up':set(), 'Dw':set()}
                    destino = 0
                    way = -1
                
                if self.andar in self.destinos[op]:
                    self.histo[self.andar] += 1
                    self.destinos[op].remove(self.andar)
                    print('{}º andar'.format(self.andar))
            
        
        self.way.pop(0)
        self.__anda__()              
            
```
